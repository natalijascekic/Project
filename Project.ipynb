## IMAGE BRIGHTNESS AND IMAGE CONTRAST
   

```python
import matplotlib.pyplot as plt 
from scipy.misc import imread,imsave
import numpy as np
import imageio
%matplotlib inline

def img(img):
    return imageio.imread(img)
```

# Image brightness

```python
def brightness(n):
    
    image = img("horse.jpg")
    
    for i in range(len(image)):
        for j in range(len(image[0])):
            for k in range(3):
                
                m = image[i][j][k] + (image[i][j][k] * (n/100))
                
                if m > 255:
                    image[i][j][k] = 255
                    
                else:
                    image[i][j][k] = m
       
                
    plt.imshow(image)
    plt.show()
    
    
```

# Image before and after the brightness

```python
plt.imshow(imageio.imread("horse.jpg"))
plt.show()
brightness(20)
```

# Image contrast

```python
def contrast(n):
    
    image = img("horse.jpg")
    
    for i in range(len(image)):
        for j in range(len(image[0])):
            for k in range(3):
                
                if image[i][j][k] < 128:
                    
                    p = image[i][j][k] - (image[i][j][k] * (n/100))
                    
                    if p < 0:
                        image[i][j][k] = 0
                        
                    else:
                        image[i][j][k] = p
                        
                else:
                    
                    m = image[i][j][k] + (image[i][j][k] * (n/100))
                    
                    if m > 255:
                        image[i][j][k] = 255
                    else:
                        image[i][j][k] = m
                                    


    plt.imshow(image)
    plt.show()
```

# Image before and after the contrast

```python
plt.imshow(imageio.imread("horse.jpg"))
plt.show()
contrast(20)
```
